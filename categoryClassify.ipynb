<h1> Title: Category Classification</h1>

<strong>Overview: In this notebook, I aim to classify categories</strong><br>
In this notebook, it covers:<br>
1.0 Zero-Shot Classification<br>
2.0 Keyword Matching<br>

```python
import pandas as pd
from transformers import pipeline
```

```python
df = pd.read_csv('Datasets/comments.csv',nrows=20)
df
```

## 1.0 Zero-Shot Classification

```python
candidate_labels = ["skincare", "makeup","hair", "other"]

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

def zero_shot_classify(text):
    result = classifier(text, candidate_labels)
    return result["labels"][0]  # take the best match

df["zero_shot_category"] = df["textOriginal"].apply(zero_shot_classify)
```

```python
print(df[["textOriginal", "zero_shot_category"]])
```

## 2.0 Keyword Matching

```python
keyword_dict = {
    "skincare": ["acne", "moisturizer", "serum", "cleanser", "toner"," beauty essence","exfoliant"],
    "makeup": ["liquid foundation", "powder foundation","concealer","blush", "foundation", "lipstick","eye","eyeshadow", "mascara", "eyeliner", "eyebrow","lip","lip gloss","lip balm"],
    "hair": ["shampoo", "conditioner","hair treatment", "hair", "curl", "straighten", "dye", "color"],
}

def keyword_match(text):
    text_lower = str(text).lower()
    for category, keywords in keyword_dict.items():
        if any(word in text_lower for word in keywords):
            return category
    return "other"

df["keyword_category"] = df["textOriginal"].apply(keyword_match)
```

```python
print(df[["textOriginal", "keyword_category"]])
```
